```python
# Dependicies
import pandas as pd
import numpy as np
from citipy import citipy

# Import file
seattle_df = pd.read_csv('data/SDOT_Collisions_-_All_Years.csv')
seattle_df.head(10)
```

```python
# Remove unnecessary columns from the dataframe
columns_to_drop = ['INCKEY', 'OBJECTID', 'COLDETKEY', 'EXCEPTRSNCODE', 'EXCEPTRSNDESC',
                   'INCDTTM', 'INTKEY', 'JUNCTIONTYPE', 'LOCATION', 'PEDROWNOTGRNT', 'REPORTNO',                   
                   'SDOT_COLDESC', 'SDOTCOLNUM', 'ST_COLDESC', 'STATUS', 'SEGLANEKEY', 'CROSSWALKKEY',                   
                   'HITPARKEDCAR', 'SOURCEDESC', 'ADDDTTM', 'MODBY', 'MODDTTM', 'ST_PARTCPNT_TYPE',                   
                   'ST_PARTCPNT_TYPE_DESC', 'ST_AGE', 'ST_CITED', 'ST_VEH_TYPE_CD', 'ST_VEH_TYPE_DESC',                   
                   'ST_INJRY_CLSS_DESC', 'SOURCE', 'ADDBY']

cleaned_seattle_df = seattle_df.drop(columns=columns_to_drop)
```

```python
# Replacing NaN values in the 'SPEEDING' column to 0
cleaned_seattle_df['SPEEDING'] = cleaned_seattle_df['SPEEDING'].replace({'NAN': np.nan, 'Y': 1}).fillna(0).astype('Int64')
```

```python
# print(cleaned_seattle_df['INATTENTIONIND'].unique())

# Replaced NaN values in 'INATTENTIONIND' column to 0
cleaned_seattle_df['INATTENTIONIND'] = cleaned_seattle_df['INATTENTIONIND'].replace({'NAN': np.nan, 'Y': 1}).fillna(0).astype('Int64')
```

```python
unique_UNDERINFL= cleaned_seattle_df['UNDERINFL'].unique()
unique_UNDERINFL
```

```python
# Reworking the multiple values in the 'UNDERINFL' column to be int form
cleaned_seattle_df['UNDERINFL'] = cleaned_seattle_df['UNDERINFL'].replace({
    'N': 0,
    '0': 0,
    'Y': 1,
    '1': 1,
    np.nan: 2
})
```

```python
unique_addrtype = cleaned_seattle_df['ADDRTYPE'].unique()
unique_addrtype
```

```python
addrtype_key = {'Block': 1, 
                'Intersection': 2, 
                'Alley': 3, 
                np.nan: 0
               }

cleaned_seattle_df['ADDRTYPE'] = cleaned_seattle_df['ADDRTYPE'].replace(addrtype_key)
```

```python
unique_collisiontype = cleaned_seattle_df['COLLISIONTYPE'].unique()
unique_collisiontype
```

```python
collisiontype_key = {'Angles': 1,
                     'Cycles': 2,
                     'Head On': 3,
                     'Left Turn': 4,
                     'Other': 5,
                     'Parked Car': 6,
                     'Pedestrian': 7,
                     'Rear Ended': 8,
                     'Right Turn': 9,
                     'Sideswipe': 10,
                     np.nan: 0
                    }

cleaned_seattle_df['COLLISIONTYPE'] = cleaned_seattle_df['COLLISIONTYPE'].replace(collisiontype_key)
```

```python
unique_lightcond = cleaned_seattle_df['LIGHTCOND'].unique()
unique_lightcond
```

```python
# Define a dictionary that maps each unique value in the 'LIGHTCOND' column to an integer key
lightcond_key = {
    'Daylight': 1,
    'Dark - Street Lights On': 2,
    'Unknown': 3,
    'Dusk': 4,
    'Dawn': 5,
    'Dark - No Street Lights': 6,
    'Other': 7,
    'Dark - Street Lights Off': 8,
    'Dark - Unknown Lighting': 9,
    np.nan: 0
}

# Use the replace() method to replace each value in the 'LIGHTCOND' column with its corresponding integer key
cleaned_seattle_df['LIGHTCOND'] = cleaned_seattle_df['LIGHTCOND'].replace(lightcond_key)

cleaned_seattle_df['LIGHTCOND'].head(10)
```

```python
unique_roadcond = cleaned_seattle_df['ROADCOND'].unique()
unique_roadcond
```

```python
roadcond_key = {'Dry': 0, 
                'Wet': 1, 
                'Unknown': 2, 
                'Snow/Slush': 3, 
                'Ice': 4, 
                'Other': 5, 
                'Standing Water': 6, 
                'Sand/Mud/Dirt': 7, 
                'Oil': 8,
                np.nan: 0
}
cleaned_seattle_df['ROADCOND'] = cleaned_seattle_df['ROADCOND'].replace(roadcond_key)
```

```python
unique_severitydesc = cleaned_seattle_df['SEVERITYDESC'].unique()
unique_severitydesc
```

```python
severitydesc_key ={
    'Property Damage Only Collision': 0, 
    'Unknown': 1, 
    'Injury Collision': 2, 
    'Serious Injury Collision': 3, 
    'Fatality Collision': 4
}

cleaned_seattle_df['SEVERITYDESC'] = cleaned_seattle_df['SEVERITYDESC'].map(severitydesc_key)
```

```python
unique_weather = cleaned_seattle_df['WEATHER'].unique()
unique_weather
```

```python
weather_key = {
    'Unknown': 0,
    'Clear': 1,
    'Partly Cloudy': 2,
    'Overcast': 3,
    'Raining': 4,
    'Snowing': 5,
    'Fog/Smog/Smoke': 6,
    'Sleet/Hail/Freezing Rain': 7,
    'Blowing Sand/Dirt': 8,
    'Severe Crosswind': 9,
    'Blowing Snow': 10,
    'Other': 11,
}

cleaned_seattle_df['WEATHER'] = cleaned_seattle_df['WEATHER'].map(weather_key)
```
